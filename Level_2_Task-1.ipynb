```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Dataset.csv')
```

```python
# Analyze the distribution of aggregate ratings and determine the most common rating range.
```

```python
agg_rating_distribution=df['Aggregate rating'].value_counts()
print(agg_rating_distribution)
```

```python
a=agg_rating_distribution.idxmax()
print('Most Common Rating Range is :', a)
```

```python
plt.hist(df['Aggregate rating'], bins=30, color='Blue', edgecolor='Black')
plt.xlabel('Aggregate rating')
plt.ylabel('count')
plt.title('Distribution of aggregate rating')
plt.show()
```

```python
#Calculate the average number of votes received by restaurants.
```

```python
avg_vote= round(df['Votes'].mean(),3)
print('The average number of votes recieved by the restaurants:',avg_vote)
```
