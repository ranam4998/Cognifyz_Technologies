```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_csv('Dataset.csv')
```

```python
print(df.columns)
```

```python
# create a histogram pr bar chart  to visualize the distribution of price ranges amont the restaurants
```

```python
df.hist(column='Price range')
plt.title('Distribution of Price Ranges')
plt.xlabel('Price range')
plt.ylabel('Counts')
plt.show()
```

```python
#Calculate the percentage of restaurants in each price range category.
```

```python
price_range_count=df['Price range'].value_counts()
total_restaurant = len(df)
percentage = round((price_range_count/total_restaurant)*100,2)
percentage_m=pd.merge(price_range_count, percentage, on='Price range')
print('The % of Restuarant in each price range category are: ',percentage_m)
```

```python
plt.pie(percentage,labels=percentage.index,autopct='%1.1f%%', colors=['Yellow', 'Green', 'Red', 'blue'])
plt.title('Price Range Percentage')
plt.show()
```
